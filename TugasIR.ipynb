```python
# ============================================================
# TUGAS INFORMATION RETRIEVAL (YUNITA ISNA D - 32602400125)
# SISTEM PENCARIAN DESTINASI WISATA KABUPATEN SEMARANG
# Metode = TF-IDF & Cosine Similarity
# ============================================================

import pandas as pd
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.metrics.pairwise import cosine_similarity

# Dataset Wisata
data = {
    "Nama Wisata": [
        "Umbul Sidomukti",
        "Candi Gedong Songo",
        "Rawa Pening",
        "Eling Bening",
        "Dusun Semilir",
        "Saloka Theme Park",
        "Kampoeng Kopi Banaran",
        "Curug Lawe",
        "Bukit Cinta Rawa Pening",
        "Museum Kereta Api Ambarawa"
    ],

    "Lokasi": [
        "Bandungan",
        "Bandungan",
        "Ambarawa",
        "Bawen",
        "Bawen",
        "Tuntang",
        "Bawen",
        "Ungaran Barat",
        "Ambarawa",
        "Ambarawa"
    ],

    "Deskripsi": [
        "Wisata alam dengan kolam renang alami, flying fox, camping, outbound dan pemandangan pegunungan.",
        "Kompleks candi Hindu di lereng Gunung Ungaran dengan panorama alam dan sejarah.",
        "Danau alami dengan wisata perahu, memancing dan panorama indah.",
        "Tempat wisata keluarga dengan restoran, kolam renang dan pemandangan Rawa Pening.",
        "Wisata keluarga dengan wahana permainan, kuliner, seluncuran warna-warni dan spot foto.",
        "Taman rekreasi terbesar di Jawa Tengah dengan berbagai wahana permainan.",
        "Wisata edukasi kopi, restoran, outbound dan kebun kopi.",
        "Air terjun alami dengan jalur trekking di kaki Gunung Ungaran.",
        "Tempat menikmati panorama Rawa Pening dari gardu pandang.",
        "Museum bersejarah yang menampilkan koleksi kereta api tua dan lokomotif uap."
    ]
}

df = pd.DataFrame(data)

df["Dokumen"] = (
    df["Nama Wisata"] + " " +
    df["Lokasi"] + " " +
    df["Deskripsi"]
)

vectorizer = TfidfVectorizer()
tfidf_matrix = vectorizer.fit_transform(df["Dokumen"])

def cari_wisata(keyword):
    query = vectorizer.transform([keyword])
    similarity = cosine_similarity(query, tfidf_matrix)
    skor = similarity.flatten()

    hasil = df.copy()
    hasil["Skor"] = skor
    hasil = hasil.sort_values(by="Skor", ascending=False)

    return hasil

# ==========================
# Input
# ==========================

print("="*60)
print(" HAI.. AKU SIPEDITA (SISTEM PENCARIAN DESTINASI WISATA KABUPATEN SEMARANG)")
print("="*60)

keyword = input("Masukkan apa yang ingin anda cari: ")

hasil = cari_wisata(keyword)

print("\nHasil Pencarian\n")

for i, row in hasil.head(3).iterrows():
    print("-"*60)
    print("Nama Wisata :", row["Nama Wisata"])
    print("Lokasi      :", row["Lokasi"])
    print("Skor        :", round(row["Skor"],3))
    print("Deskripsi   :", row["Deskripsi"])

print("-"*60)
print("Pencarian selesai.")
```
